## Load the helpers

```python
# Drives the functions in pull_statcan.py (one source of truth).
from pull_statcan import TABLES, list_tables, check, pull_one, DATA
print('Loaded. Configured tables:', list(TABLES))
```

```python
list_tables()
```

## 2. Freshness check — which tables changed today?

```python
check()
```

## 3. Download every table
Writes each table + a provenance JSON into `data/raw/<domain>/`. Re-run only when step 2 flags a change.

```python
for key in TABLES:
    pull_one(key)
```

## 4. Confirm what landed

```python
for f in sorted((DATA / 'raw').glob('**/*.csv')):
    if f.name.endswith('_metadata.csv'):
        continue
    rows = sum(1 for _ in f.open(encoding='utf-8')) - 1
    print(f'{f.relative_to(DATA)}  ->  {rows:,} rows')
```
